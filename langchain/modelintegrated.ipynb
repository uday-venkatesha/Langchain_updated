### Streaming And Batch

### Streaming
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses.
Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time:

```python
model.invoke("write a 200 lines paragraph on AI")    
```

```python

for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text, end="|", flush=True)
```

```python
# stream and print the output in 200-word segments
prompt = "Why do parrots have colorful feathers?"
words = []
segment_index = 1

for chunk in model.stream(prompt):
    text = chunk.text or ""
    if not text:
        continue
    words.extend(text.split())
    while len(words) >= 200:
        segment = " ".join(words[:200])
        print(f"--- segment {segment_index} ---\n{segment}\n", flush=True)
        segment_index += 1
        words = words[200:]

# print any remaining words (<200) as the final segment
if words:
    print(f"--- segment {segment_index} (final) ---\n{' '.join(words)}\n", flush=True)
```
